# Preparación de alcohol al 70% desde distintas concentraciones

Calcular la concentración porcentual de una solución de alcohol para desinfectar requiere considerar
la concentración del alcohol de partida (v/v).

La fórmula base es la **ecuación de dilución**:

$$C_1 \cdot V_1 = C_2 \cdot V_2$$

Donde:
- $C_1$ = concentración inicial del alcohol disponible (ej: 95%, 96%, 99.9%)
- $V_1$ = volumen de alcohol inicial que se usará (mL)
- $C_2$ = concentración final deseada (70%)
- $V_2$ = volumen total de la solución final (mL)

A partir de esto se calcula cuánta agua agregar: $\text{agua} = V_2 - V_1$

```python
def solucion(vol_alcohol, concentracion_inicial, concentracion_final=70):
    """
    Calcula cuánta solución final se obtiene y cuánta agua agregar.

    Parámetros:
    -----------
    vol_alcohol          : float  -> mL de alcohol disponibles
    concentracion_inicial: float  -> % del alcohol de partida (ej: 95, 96, 99.9)
    concentracion_final  : float  -> % deseado en la solución final (por defecto 70)

    Retorna:
    --------
    tupla (vol_final, vol_agua) en mL (redondeados)
    """
    if concentracion_inicial <= 0 or concentracion_inicial > 100:
        raise ValueError("La concentración inicial debe estar entre 0 y 100%.")
    if concentracion_final >= concentracion_inicial:
        raise ValueError("La concentración final debe ser menor que la inicial.")

    # Usando C1*V1 = C2*V2  =>  V2 = (C1*V1) / C2
    vol_final = (concentracion_inicial * vol_alcohol) / concentracion_final
    vol_agua  = vol_final - vol_alcohol

    print(f"Con {vol_alcohol} mL de alcohol al {concentracion_inicial}%:")
    print(f"  Agregar  : {round(vol_agua)} mL de agua")
    print(f"  Obtiene  : {round(vol_final)} mL de solución al {concentracion_final}%")
    print()

    return round(vol_final), round(vol_agua)
```

## Ejemplos con 325 mL de alcohol de distintas concentraciones

```python
vol = 325  # mL de alcohol disponibles

solucion(vol, 95)
solucion(vol, 96)
solucion(vol, 99.9)
```

## Calcular para cualquier volumen y concentración

```python
# Modificar estos valores según lo que tengas disponible
vol_disponible      = 500   # mL de alcohol que tienes
concentracion_stock = 96    # % de ese alcohol

solucion(vol_disponible, concentracion_stock)
```
