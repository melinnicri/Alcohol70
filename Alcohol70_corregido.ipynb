# Preparación de alcohol al 70% desde distintas concentraciones

Para preparar una solución desinfectante al 70% v/v, se aplica la **ecuación de dilución**:

$$V_1 \times C_1 = V_2 \times C_2$$

Donde:
- $C_1$ = concentración del alcohol disponible (95%, 96% o 99.9%)
- $V_1$ = volumen de alcohol inicial a usar (lo que queremos calcular)
- $C_2$ = concentración final deseada (70%)
- $V_2$ = volumen total de solución final que queremos preparar

Despejando $V_1$:

$$V_1 = \frac{V_2 \times C_2}{C_1}$$

El agua a agregar es simplemente:

$$V_{agua} = V_2 - V_1$$

```python
def preparar_solucion(vol_final, concentracion_inicial, concentracion_final=70):
    """
    Calcula cuánto alcohol y agua se necesitan para preparar
    una solución de concentración deseada.

    Parámetros:
    -----------
    vol_final            : float -> volumen de solución final a preparar (mL)
    concentracion_inicial: float -> % del alcohol de partida (ej: 95, 96, 99.9)
    concentracion_final  : float -> % deseado en la solución final (por defecto 70)

    Retorna:
    --------
    tupla (vol_alcohol, vol_agua) en mL (redondeados a 1 decimal)
    """
    if concentracion_inicial <= 0 or concentracion_inicial > 100:
        raise ValueError("La concentración inicial debe estar entre 0 y 100%.")
    if concentracion_final >= concentracion_inicial:
        raise ValueError("La concentración final debe ser menor que la inicial.")

    # V1 = (V2 * C2) / C1
    vol_alcohol = (vol_final * concentracion_final) / concentracion_inicial
    vol_agua    = vol_final - vol_alcohol

    print(f"Para preparar {vol_final} mL de solución al {concentracion_final}% "
        f"desde alcohol al {concentracion_inicial}%:")
    print(f"  Alcohol al {concentracion_inicial}%: {round(vol_alcohol, 1)} mL")
    print(f"  Agua                : {round(vol_agua, 1)} mL")
    print()

    return round(vol_alcohol, 1), round(vol_agua, 1)
```

## Para 500 mL de solución al 70%

```python
preparar_solucion(500, 95)
preparar_solucion(500, 96)
preparar_solucion(500, 99.9)
```

## Para 1000 mL de solución al 70%

```python
preparar_solucion(1000, 95)
preparar_solucion(1000, 96)
preparar_solucion(1000, 99.9)
```

## Calcular para cualquier volumen y concentración

```python
# Modificar estos valores según lo que necesites
vol_a_preparar      = 750   # mL de solución final que quieres preparar
concentracion_stock = 96    # % del alcohol que tienes disponible

preparar_solucion(vol_a_preparar, concentracion_stock)
```

## Tabla resumen: 500 y 1000 mL para los tres alcoholes

```python
volumenes       = [500, 1000]
concentraciones = [95, 96, 99.9]

print(f"{'Vol. final (mL)':>15} | {'Conc. inicial (%)':>17} | {'Alcohol a usar (mL)':>19} | {'Agua (mL)':>9}")
print("-" * 68)

for vol in volumenes:
    for conc in concentraciones:
        valc = round((vol * 70) / conc, 1)
        vagu = round(vol - valc, 1)
        print(f"{vol:>15} | {conc:>17} | {valc:>19} | {vagu:>9}")
    print("-" * 68)
```
